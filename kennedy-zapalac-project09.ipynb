## Question 1

a. Write a query to find out who won the 2022 World Baseball Series from table `seriespost`?

b. For this champion team, please find out the home runs (hr) rate and runs batted in (rbi) rate in year 2022 from table batting. Round the rates to 2 decimals. You may get rates by

    - hr_rate = sum of home runs / sum of hits
    - rbi_rate = sum of runs batted in / sum of hits

```python
%%sql
SELECT * FROM seriespost WHERE yearID = 2022 AND round = 'WS';
```

```python
%%sql
SELECT 
    ROUND(CAST(SUM(HR) AS REAL)/SUM(H), 2) AS hr_rate, # must cast at least one of the inputs as a real number so that the output is a decimal rather than 0
    ROUND(CAST(SUM(RBI) AS REAL)/SUM(H), 2) AS rbi_rate 
FROM 
    batting 
WHERE 
    yearID = 2022 AND teamID = 'HOU'; 
```

The Houston team won the 2022 world series and had a home run rate of 0.16 and a runs batted in rate of 0.53.

## Question 2
a. For the champion team from question 1, please write a query that counts the number of RBIs for each athlete in the champion team during year 2022, using the `batting` table. Display your output in ascending order.

b. Run the query again, but this time, display the output in descending order.

c. Which athlete has the highest RBIs in this question? Please provide the playerâ€™s `playerID`, along with their first name and last name, from the `people` table

```python
%%sql
SELECT
    b.playerID,
    p.nameFirst,
    p.nameLast,
    b.RBI
FROM
    batting AS b
JOIN
    people AS p
ON
    b.playerID = p.playerID
WHERE
    teamID = 'HOU' AND yearID = 2022
ORDER BY
    b.RBI ASC;
```

## Question 4
a. Write a query that gets the average salary for each athlete in the database. Display your output in descending order. Limit the output to 100 rows (i.e., to the top 100 salaries).

```python
%%sql
SELECT
    playerID,
    AVG(salary) as avg_salary
FROM
    salaries
GROUP BY
    playerID
ORDER BY
    avg_salary DESC
LIMIT 100;
```

## Question 5

Now create your own query about a topic that you are interested in. Use at least one of the aggregation functions, such as min, max, count, or sum. Be sure to use group by and display the results in order with order by

My question: which team has the highest average home runs over the years?

```python
%%sql
SELECT
    name,
    AVG(HR) as avg_hr
FROM 
    teams
GROUP BY
    name
ORDER BY
    avg_hr DESC
LIMIT 1;
```

The Colorado Rockies have the highest average home runs over the years.
